```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO  # Importar StringIO desde io

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
df = pd.read_csv(StringIO(data))

# Convertir la columna 'fecha' a formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])  # Corregir aqu

# Grfico de lneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o')  # Aadir marcadores para mejor visualizacin
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid()
plt.show()

# Crear un grfico de dispersin
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='orange')  # Cambiar color para diferenciar
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid()
plt.show()

# Crear un histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribucin de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid()
plt.show()
```
